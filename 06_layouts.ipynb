### Columns

```python
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create a new plot
s1 = plotting.figure(plot_width=250, plot_height=250, title=None)
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

# create another one
s2 = plotting.figure(plot_width=250, plot_height=250, title=None)
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)

# create and another
s3 = plotting.figure(plot_width=250, plot_height=250, title=None)
s3.square(x, y2, size=10, color="olive", alpha=0.5)

# put the results in a column and show
plotting.show(layouts.column(s1, s2, s3))
```

```python
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create a new plot
s1 = plotting.figure(plot_width=450, plot_height=250, title=None)
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

# create another one
s2 = plotting.figure(plot_width=450, plot_height=250, title=None)
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)

# create and another
s3 = plotting.figure(plot_width=450, plot_height=250, title=None)
s3.square(x, y2, size=10, color="olive", alpha=0.5)

# create and another
s4 = plotting.figure(plot_width=450, plot_height=250, title=None)
s4.square(x, y2, size=10, color="olive", alpha=0.5)

# put the results in a row
plotting.show(layouts.column(layouts.row(s1, s2), layouts.row(s3, s4)))
```

### Grid Layouts

```python
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create three plots
p1 = plotting.figure(title=None)
p1.circle(x, y0, size=10, color=palettes.Viridis3[0])
p2 = plotting.figure(title=None)
p2.triangle(x, y1, size=10, color=palettes.Viridis3[1])
p3 = plotting.figure(title=None)
p3.square(x, y2, size=10, color=palettes.Viridis3[2])

# make a grid
grid = plotting.gridplot([p1, p2, p3], ncols=2, plot_width=450, plot_height=250)

# show the results
plotting.show(grid)
```
