### Basic Interactors

```python
x = np.linspace(0, 2*np.pi, 2000)
y = np.sin(x)
```

```python
p = plotting.figure(title="simple line example", plot_height=300, plot_width=600, y_range=(-5,5))
r = p.line(x, y, color="#2222aa", line_width=1)
```

```python
def update(f, w=1, A=1, phi=0):
    if   f == "sin": func = np.sin
    elif f == "cos": func = np.cos
    elif f == "tan": func = np.tan
    r.data_source.data['y'] = A * func(w * x + phi)
    io.push_notebook()
```

```python
plotting.show(p, notebook_handle=True)
```

```python
ipywidgets.interact(update, f=["sin", "cos", "tan"], w=(0,100), A=(1,5), phi=(0, 20, 0.1))
```

### Linking Plots

```python
x = list(range(11))
y0 = x
y1 = [10-xx for xx in x]
y2 = [abs(xx-5) for xx in x]

# create a new plot
s1 = plotting.figure(plot_width=320, plot_height=320, title=None)
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

# create a new plot and share both ranges
s2 = plotting.figure(plot_width=320, plot_height=320, x_range=s1.x_range, y_range=s1.y_range, title=None)
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)

# create a new plot and share only one range
s3 = plotting.figure(plot_width=320, plot_height=320, x_range=s1.x_range, title=None)
s3.square(x, y2, size=10, color="olive", alpha=0.5)

p = layouts.gridplot([[s1, s2, s3]])

# show the results
plotting.show(p)
```

### Hiding Glyphs

```python
from bokeh.sampledata.stocks import AAPL, IBM, MSFT, GOOG

p = plotting.figure(plot_width=900, plot_height=250, x_axis_type="datetime")
p.title.text = 'Click on legend entries to hide the corresponding lines'

for data, name, color in zip([AAPL, IBM, MSFT, GOOG], ["AAPL", "IBM", "MSFT", "GOOG"], palettes.Spectral4):
    df = pd.DataFrame(data)
    df['date'] = pd.to_datetime(df['date'])
    p.line(df['date'], df['close'], line_width=2, color=color, alpha=0.8, legend=name)

p.legend.location = "top_left"
p.legend.click_policy="hide"

plotting.show(p)
```
